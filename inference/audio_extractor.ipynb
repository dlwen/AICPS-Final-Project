```python
from moviepy.editor import VideoFileClip
import os

def convert_mp4_to_mp3(input_path, output_path=None):
    """
    Convert MP4 video file to MP3 audio file.

    Args:
        input_path (str): Path to the input MP4 file
        output_path (str, optional): Path to save the output MP3 file.
                                   If not provided, will use the same name as input with .mp3 extension
    """
    try:
        # Load the video file
        video = VideoFileClip(input_path)

        # If output path is not provided, create one with the same name
        if output_path is None:
            output_path = os.path.splitext(input_path)[0] + '.mp3'

        # Extract audio and save as MP3
        video.audio.write_audiofile(output_path)

        print(f"Successfully converted {input_path} to {output_path}")

    except Exception as e:
        print(f"Error during conversion: {str(e)}")
    finally:
        # Close the video file
        if 'video' in locals():
            video.close()

```

```python
import os

folder_path = "data/video"

for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    print(filename)
    output_path = "data/audio/" + filename.split(".")[0] + ".mp3"
    print(output_path)
    convert_mp4_to_mp3(file_path, output_path)
```
