```python
import os
import whisper

folder_path = "data/video"
confidence = []
model = whisper.load_model("turbo")

for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    print("=============")
    print(filename)
    output_path = "data/ASR_result/" + filename.split(".")[0] + ".txt"
    print(output_path)

    result = model.transcribe(file_path)
    predicted = result["text"]
    with open(output_path, 'x') as file:
      file.write(predicted)

    confidence_by_segment = []
    for segment in result["segments"]:
      confidence_by_segment.append(segment["avg_logprob"])

    confidence.append(sum(confidence_by_segment) / len(confidence_by_segment))
    print(predicted)
    print(confidence_by_segment)
```
